```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Zomato data .csv")
```

```python
df.head()
```

```python
print(df)
```

# covert the data type of colulmn- rate

```python
def handlerate(value):
    value = str(value).split('/')
    value = value[0]   
    return float(value)

df['rate'] = df['rate'].apply(handlerate)
print(df.head())
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("type of resturant")
```

# conclusion - majority of the resturant falls in dinning category

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

```python
df.head()
```

```python
plt.hist(df['rate'],bins =7,color='lightblue')
plt.title("ratings distribution")
plt.show()
```

```python
couple_data=df['approx_cost(for two people)']
sns.countplot(x=couple_data, palette="Set2")
```

# which mode revieve highest rating

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = df,color ='pink')
```

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
